```python
import pandas as pd
```

```python
df = pd.read_excel("Slate Robotics 2022 Data.xlsx")
```

```python
df1 = df[['First', 'Last', 'Field:Parent1_First', 'Field:Parent1_Last', 'Field:Parent1_Phone']].copy()
```

```python
df1
```

```python
df1['Camper Name'] = df1['First'] + " " + df1['Last']
df1['Parent Name'] = df1['Field:Parent1_First'] + " " + df1['Field:Parent1_Last']
df1.rename(columns = {'Field:Parent1_Phone':'Parent Phone'}, inplace = True)
```

```python
df1.drop(['First', 'Last', 'Field:Parent1_First', 'Field:Parent1_Last'], axis = 1)
new_cols = ["Camper Name","Parent Name","Parent Phone"]
df1=df1[new_cols]
df1=df1.reindex(columns=new_cols)
```

```python
df1.to_excel("CamperParentPhoneNumbers.xlsx")
```
