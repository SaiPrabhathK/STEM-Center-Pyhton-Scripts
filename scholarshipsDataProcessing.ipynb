```python
import pandas as pd
```

```python
#### Reading Spreadsheets of camps into respective data frames
# Scholarships Forms Data
ScholarshipsForms = pd.read_excel("scholarships forms data.xlsx")
# Explosives Camp Session I
Explosives_Session_I = pd.read_excel("explosives i.xlsx")
# Explosives Camp Session II
Explosives_Session_II = pd.read_excel("explosives ii.xlsx")
# Robotics Camp
Robotics = pd.read_excel("robotics.xlsx")
# Jackling Session I
Jackling_Session_I = pd.read_excel("jackling session i.xlsx")
# Jackling Session II
Jackling_Session_II = pd.read_excel("jackling session ii.xlsx")
# Jackling Session III
Jackling_Session_III = pd.read_excel("jackling session iii.xlsx")
# Art of Science
Art_of_Science = pd.read_excel("art of science.xlsx")
# Nuclear Engineering Camp
Nuclear_Engineering = pd.read_excel("nuclear.xlsx")
# Cyberminer Camp
Cyberminer = pd.read_excel("cyberminer.xlsx")
# Space - The Final Frontier Camp
Space = pd.read_excel("space.xlsx")
# Aerospace Camp
Aerospace = pd.read_excel("aerospace.xlsx")
# Formula SAE
Formula_SAE = pd.read_excel("formula sae.xlsx")
# Subsurface Camp
Subsurface = pd.read_excel("subsurface.xlsx")
# Materials Camp is not included since it's a sponsored (free) camp

campNamesList = ["Explosives_Session_I", "Explosives_Session_II", "Robotics", "Jackling_Session_I", "Jackling_Session_II", "Jackling_Session_III", "Art_of_Science", "Nuclear_Engineering",
                 "Cyberminer", "Space", "Aerospace", "Formula_SAE", "Subsurface"]
def fullName(frst, lst):
  return (frst.strip().capitalize() + " " + lst.strip().capitalize())
```

```python
names = list()
attendeesinEachCamp = list()
namesInEachCamp = list()
for i in range(0, len(ScholarshipsForms)):
  names.append(fullName(ScholarshipsForms["First"][i], ScholarshipsForms["Last"][i]))
for i in campNamesList:
  #print(i)
  #count = 0
  for j in range (0, len(globals()[i].index)): ### ********* ###
    #print(j)
    x = fullName(globals()[i]["First"][j], globals()[i]["Last"][j])
    if (x) in names:
      #print(x)
      #count = count + 1
      attendeesinEachCamp.append(i)
      namesInEachCamp.append(x)
  #print(count)
#names
```

```python
data = []
data.append(namesInEachCamp)
data.append(attendeesinEachCamp)
data
```
