```python
from PIL import Image, ImageDraw, ImageFont
import pandas as pd
from google.colab import files
form = pd.read_csv('Robotics Camp 2022.csv')
```

```python
firstName_list = form['First'].to_list()
lastName_list = form['Last'].to_list()
```

```python
for i in range(0, len(firstName_list)):
    fullName = firstName_list[i] + " " + lastName_list[i]
    print(fullName)
    im = Image.open("Robotics certificate frame.png")
    d = ImageDraw.Draw(im)
    firstNameLocation = (220, 1040)
    #lastNameLocation = (353, 1040)
    text_color = (0, 0, 0)
    if (i != 9):
      font = ImageFont.truetype("Forte.ttf", 150)
    else:
      font = ImageFont.truetype("Forte.ttf", 100)
    d.text(firstNameLocation, fullName, fill=text_color,font=font)
    rgb_im = im.convert('RGB')
    rgb_im.save("Final Certificates/"+fullName+".pdf")
    files.download("Final Certificates/"+fullName+".pdf")
```
