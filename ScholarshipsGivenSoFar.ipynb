```python
#### Importing all camps' data from Slate
# Explosives Camp Session I
Explosives_Session_I = pd.read_excel("explosives i.xlsx")
# Explosives Camp Session II
Explosives_Session_II = pd.read_excel("explosives ii.xlsx")
# Robotics Camp
Robotics = pd.read_excel("robotics.xlsx")
# Jackling Session I
Jackling_Session_I = pd.read_excel("jackling session i.xlsx")
# Jackling Session II
Jackling_Session_II = pd.read_excel("jackling session ii.xlsx")
# Jackling Session III
Jackling_Session_III = pd.read_excel("jackling session iii.xlsx")
# Art of Science
Art_of_Science = pd.read_excel("art of science.xlsx")
# Nuclear Engineering Camp
Nuclear_Engineering = pd.read_excel("nuclear.xlsx")
# Cyberminer Camp
Cyberminer = pd.read_excel("cyberminer.xlsx")
# Space - The Final Frontier Camp
Space = pd.read_excel("space.xlsx")
# Aerospace Camp
Aerospace = pd.read_excel("aerospace.xlsx")
# Formula SAE
Formula_SAE = pd.read_excel("formula sae.xlsx")
# Materials Camp not included since it is sponsored (free) camp & Subsurface Camp is also not included.

campNamesList = ["Explosives_Session_I", "Explosives_Session_II", "Robotics", "Jackling_Session_I", "Jackling_Session_II", "Jackling_Session_III", "Art_of_Science", "Nuclear_Engineering",
                 "Cyberminer", "Space", "Aerospace", "Formula_SAE"]
#d = globals()
#campNamesList = [v for k, v in d.items() if isinstance(v, pd.DataFrame)]
#campNamesList
```

```python
names = list()
campsAttending = list()
campCost = list()
cost = 0
for i in campNamesList:
  #print(i)
  for j in range (0, len(globals()[i].index)): ### ********* ###
    #print(j)
    formattedName = ""
    if pd.notna(globals()[i]["Notes"][j]):
      if "received scholarship" in globals()[i]["Notes"][j]:
        names.append(globals()[i]["First"][j].capitalize() + " " + globals()[i]["Last"][j].capitalize())
        campName = i.split("_")
        for k in campName:
          formattedName = formattedName + " " + k
        campsAttending.append(formattedName.strip())
        note = globals()[i]["Notes"][j].split()
        print(note[2])
        campCost.append(note[2])
        cost = cost + int(note[2])
        #print(globals()[i]["Notes"][j])
print(campCost, names, campsAttending)
print(cost)
```

```python
data = []
data.append(names)
data.append(campsAttending)
data.append(campCost)
data
```

```python
df = pd.DataFrame(data).transpose()
df.columns=['Student Names', 'Camp Attending', 'Camp Cost']
```
