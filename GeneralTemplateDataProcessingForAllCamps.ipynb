```python
import pandas as pd
```

```python
########################## Always put the latest form here (you can change the name as you want, make sure it matches the downloaded spread sheet)###########################
formData = pd.read_excel("MS&T Summer Camp Registrants 2022 Form (Responses).xlsx")
```

```python
#options = ['Robotics Camp 2022'] 
#df = formData[formData['Summer Camp(s) Attending'].isin(options)] 
#df = formData[formData['Summer Camp(s) Attending'] == "Robotics Camp 2022"]
#formData
```

```python
#campList = list()
multipleCampIndexes = list()
c = 0
for i in range(0, len(formData)):
  camps = formData['Summer Camp(s) Attending'][i].split(', ')
  #print(i)
  #print(camps)
  for j in camps:
    #print(camps)
    if (j == 'Formula SAE Electric Car Camp 2022'): ################################ Change Camp Name Here! #################################
      multipleCampIndexes.append(i)
      c+=1
      
      #print(c)
#c, multipleCampIndexes
```

```python
#df = formData.loc[[multipleCampIndexes[55]]]
campDf = pd.DataFrame()
for i in multipleCampIndexes:
  campDf = campDf.append(formData.loc[[i]])
#campDf.columns
#campDf[['Student First Name', '1. Will your child be bringing any type of medicine to this event?']]
```

```python
medicationDf = pd.DataFrame()
avConsentDf = pd.DataFrame()
allergiesDf = pd.DataFrame()
specialDietaryNeedsDf = pd.DataFrame()
camperParentPhoneDf = pd.DataFrame()
signInOutDf = pd.DataFrame()
```

```python
for i in multipleCampIndexes:
  if(campDf['1. Will your child be bringing any type of medicine to this event?'][i] == "Yes"):
    medicationDf = medicationDf.append(campDf.loc[[i]])
  if(campDf['I consent to the Audio/Visual Release Form'][i] == "No"):
    avConsentDf = avConsentDf.append(campDf.loc[[i]])
  if(campDf['2. Does your child have any allergies?`'][i] == "Yes"):
    allergiesDf = allergiesDf.append(campDf.loc[[i]])
  if(pd.notna(campDf['4. Does your child have any special dietary needs?'][i])):
    specialDietaryNeedsDf = specialDietaryNeedsDf.append(campDf.loc[[i]])
#medicationDf
```

```python
columnsInMedicationDf = ['Student Last Name', 'Student First Name', 'Student Middle Name', '1. Will your child be bringing any type of medicine to this event?', 'If yes, give type and instructions', 
                         'Name of Medication #1', 'Is medication #1 prescribed?', 'Days to be given for medication #1', 'Times to be given for medication #1', 'Amount to be given for medication #1', 
                         'Doses of Medication Required for #1', 'The following Over-the-Counter (OTC) medications will be available from the camp nurse or camp director/coordinator. These medications will be dispensed as labeled on the original container. Please indicate all, if any, that may be given to your child, on an as needed basis:', 
                         '5. Does your child have any other restrictions or needs, not described above?', 'Other information']
```

```python
for i in medicationDf.columns:
  if( i not in columnsInMedicationDf):
    #print (i)
    del medicationDf[i]
#medicationDf
```

```python
columnsInAVConsentDf = ['Student Last Name', 'Student First Name', 'Student Middle Name', 'I consent to the Audio/Visual Release Form']
for i in avConsentDf.columns:
  if( i not in columnsInAVConsentDf):
    #print (i)
    del avConsentDf[i]
#avConsentDf.size
```

```python
columnsInAllergiesDf = ['Student Last Name', 'Student First Name', 'Student Middle Name', '2. Does your child have any allergies?`', 'If yes, explain', '4. Does your child have any special dietary needs?']
for i in allergiesDf.columns:
  if( i not in columnsInAllergiesDf):
    del allergiesDf[i]
#allergiesDf
```

```python
columnsInSpecialDietaryNeedsDf = ['Student Last Name', 'Student First Name', 'Student Middle Name', '4. Does your child have any special dietary needs?']
for i in specialDietaryNeedsDf.columns:
  if(i not in columnsInSpecialDietaryNeedsDf):
    del specialDietaryNeedsDf[i]
#specialDietaryNeedsDf
```

```python
df = pd.read_excel("Slate Formula SAE 2022 Data.xlsx")
df1 = df[['First', 'Last', 'Mobile', 'Field:Parent1_First', 'Field:Parent1_Last', 'Field:Parent1_Phone']].copy()
df1['Camper Name'] = df1['First'] + " " + df1['Last']
df1['Parent Name'] = df1['Field:Parent1_First'] + " " + df1['Field:Parent1_Last']
df1.rename(columns = {'Field:Parent1_Phone':'Parent Phone', 'Mobile':'Camper Phone'}, inplace = True)
df1.drop(['First', 'Last', 'Field:Parent1_First', 'Field:Parent1_Last'], axis = 1)
new_cols = ["Camper Name", "Camper Phone","Parent Name","Parent Phone"]
df1 = df1[new_cols]
df1 = df1.reindex(columns=new_cols)
```

```python
camperParentPhoneDf = df1.copy()
```

```python
signInOutDf = campDf.copy()
columnsInSignInOutDf = ['Student Last Name', 'Student First Name', 'Parent/Guardian Last Name', 'Parent/Guardian First Name', 'Are there other authorized adults that are allowed to pick up your child or receive information, other than the person completing this form? Please list their names and contact information below. ']
for i in signInOutDf.columns:
  if(i not in columnsInSignInOutDf):
    del signInOutDf[i]
signInOutDf['Parent Signature Check In-Date'] = ''
signInOutDf['Parent Signature Check Out-Date'] = ''
signInOutDf
```

```python
with pd.ExcelWriter("Formula 2022 Processed Sheet.xlsx") as writer:
   
    # use to_excel function and specify the sheet_name and index
    # to store the dataframe in specified sheet
    campDf.to_excel(writer, sheet_name="Full Form", index=False)
    medicationDf.to_excel(writer, sheet_name="Medication(Yes)", index=False)
    if (avConsentDf.size != 0):
      avConsentDf.to_excel(writer, sheet_name="AV Consent(No)", index=False)
    allergiesDf.to_excel(writer, sheet_name="Allergies", index=False)
    specialDietaryNeedsDf.to_excel(writer, sheet_name = "Special Dietary Needs", index = False)
    camperParentPhoneDf.to_excel(writer, sheet_name = "Camper Parent Phone", index = False)
    signInOutDf.to_excel(writer, sheet_name = "Sign In Sign Out", index = False)
```
