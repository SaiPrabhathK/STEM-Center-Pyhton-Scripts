```python
import pandas as pd
```

```python
slateData = pd.read_excel("Slate Subsurface Engineering Camp 2022 Data.xlsx")
googleFormData = pd.read_excel("MS&T Summer Camp Registrants 2022 Form (Responses).xlsx")
```

```python
multipleCampIndexes = list()
c = 0
for i in range(0, len(googleFormData)):
  camps = googleFormData['Summer Camp(s) Attending'][i].split(', ')
  #print(i)
  #print(camps)
  for j in camps:
    #print(camps)
    if (j == 'Subsurface Engineering Camp (Day camp)'): ################################ Change Camp Name Here! #################################
      multipleCampIndexes.append(i)
      c+=1
c
```

```python
campDf = pd.DataFrame()
for i in multipleCampIndexes:
  campDf = campDf.append(googleFormData.loc[[i]])
campDf.to_excel("Space Campers.xlsx")
```

```python
campData = pd.read_excel("Space Campers.xlsx")
#campData.columns, slateData.columns
```

```python
firstNamesInCampData = campData['Student First Name'].copy().to_list()
lastNamesInCampData = campData['Student Last Name'].copy().to_list()
```

```python
#firstNamesInCampData, lastNamesInCampData
```

```python
firstNamesInSlateData = slateData['First'].copy().to_list()
lastNamesInSlateData = slateData['Last'].copy().to_list()
```

```python
#firstNamesInSlateData, lastNamesInSlateData
```

```python
namesInCampData = list()
for i in range(0, len(firstNamesInCampData)):
  namesInCampData.append(firstNamesInCampData[i].strip().capitalize() + ' ' + lastNamesInCampData[i].strip().capitalize())
namesInCampData = sorted(namesInCampData)
print(namesInCampData)
```

```python
namesInSlateData = list()
for i in range(0, len(firstNamesInSlateData)):
  namesInSlateData.append(firstNamesInSlateData[i].strip().capitalize() + ' ' + lastNamesInSlateData[i].strip().capitalize())
namesInSlateData = sorted(namesInSlateData)
print(namesInSlateData)
```

```python
namesNotInGoogleForm = list()
for i in namesInSlateData:
  if i not in namesInCampData:
    namesNotInGoogleForm.append(i)
namesNotInGoogleForm = sorted(namesNotInGoogleForm)
print(namesNotInGoogleForm)
df = pd.DataFrame()
df['Names'] = pd.DataFrame(namesNotInGoogleForm)
df['Reason'] = ""
df
```

```python
df.to_excel("NamesNotInGoogleForms.xlsx")
```
