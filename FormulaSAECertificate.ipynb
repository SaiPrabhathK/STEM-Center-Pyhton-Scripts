```python
from PIL import Image, ImageDraw, ImageFont
import PIL
import pandas as pd
from google.colab import files
form = pd.read_excel('Slate Formula SAE 2022 Data.xlsx')
```

```python
firstName_list = form['First'].to_list()
lastName_list = form['Last'].to_list()
#firstName_list, lastName_list
```

```python
for i in range(0, len(firstName_list)):
#for i in range(0, 1):
    fullName = firstName_list[i].strip() + " " + lastName_list[i].strip()
    #print(fullName)
    fullName = fullName.title()
    im = Image.open("Formula Template.jpg").convert('RGB')
    W, H = im.size
    print(W, H)
    H = 550
    d = ImageDraw.Draw(im)
    #print(wf,h)
    
    #print(w)
    #firstNameLocation = (450, 950)
    #lastNameLocation = (353, 1040)
    text_color = PIL.ImageColor.getrgb("black")
    font = ImageFont.truetype("agency-fb-12.ttf", 250)
    w, h = d.textsize(fullName,font)
    print((W/2 - w/2))
    d.text(((W/2 - w/2),H), fullName, fill=text_color, font=font)
    rgb_im = im.convert('RGB')
    #rgb_im.save(fullName+".pdf")
    rgb_im.save("Final Certificates/"+fullName+".pdf")
    files.download("Final Certificates/"+fullName+".pdf")
```
